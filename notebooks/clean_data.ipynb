```python
import pandas as pd
import csv

file_path = '../SmartCalorie_AI/data/data.csv'
output_path = '../SmartCalorie_AI/data/data_cleaned.csv'

cleaned_rows =[]
with open(file_path, 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        if len(row) >= 12:
            cleaned_rows.append(row[:12])

df = pd.DataFrame(cleaned_rows[1:], columns=cleaned_rows[0])

print(f"Data loaded successfully. Shape: {df.shape}")

df = df[df['user_id'] != 'user_id']

# 4. Convert the math columns back into actual numbers
numeric_cols =['age', 'height_cm', 'weight_kg', 'duration_minutes', 'avg_heart_rate', 'calories_burned']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 5. Drop any rows that completely broke during number conversion, just to be safe
df = df.dropna(subset=['calories_burned'])

# 6. Reset the index so the row numbers are perfect
df = df.reset_index(drop=True)

print(f"Cleaned Data Shape: {df.shape}")

# 7. Save the cleaned dataset
df.to_csv(output_path, index=False)

print(f"Success! Cleaned data saved to: {output_path}")

# Display the first few rows
df.head()
```
